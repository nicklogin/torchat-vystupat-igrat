```python
from lxml import etree
```

```python
corpora = etree.fromstring(open('annot.opcorpora.no_ambig.nonmod.xml', 'rb').read())
```

```python
sents = []


for sentence in corpora.xpath('//tokens'):
    sent = []
    for token in sentence.xpath('token'):
        wordform = token.xpath('@text')
        lemma = token.xpath('tfr/v/l/@t')[0]
        #print(lemma)
        sent.append((wordform,lemma))
    sents.append(sent)
```

```python
from ufal.udpipe import Model, Pipeline
```

```python
model = Model.load('russian-syntagrus-ud-2.3-181115.udpipe')
```

```python
pipeline = Pipeline(model, 'tokenize', Pipeline.DEFAULT, Pipeline.DEFAULT, 'epe')
```

```python
pipeline.process("Мама играла драму. Маша ела кашу")
```

```python
import re
```

```python
re.sub('\[.*?\]', '', 'a[bc]def[] g[hijk]lmn')
```

```python
def process_context_list(context_list, pipeline, item, outp_file):
    outp = []
    for index, row in context_list.iterrows():
        #line = row['Left context'].strip()+' '+row['Center'].strip()+row['Punct'].strip()+' '+row['Right context'].strip()
        line = row['Full context'].strip()
        text = re.sub('\[.*?\]', '', line)
        parses = [json.loads(sent) for sent in pipeline.process(text).splitlines()]
        tokens = [node['properties']['lemma']+'_'+node['properties']['upos'] for parse in parses for node in parse["nodes"] if node['properties']['lemma']+'_'+node['properties']['upos']!=item]
        outp.append((tokens, line))
    with open(outp_file, 'w', encoding='utf-8') as f:
        json.dump(outp, f, ensure_ascii=False)
```

```python
process_context_list(играть, pipeline, 'играть_VERB','igrat.json')
process_context_list(выступать, pipeline, 'выступать_VERB', 'vystupat.json')
process_context_list(торчать, pipeline, 'торчать_VERB', 'torchat.json')
```
